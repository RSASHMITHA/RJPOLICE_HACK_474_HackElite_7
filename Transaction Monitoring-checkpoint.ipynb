```python
import pandas as pd
data = pd.read_csv("data.csv")
data
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_FraudIndicator = data.loc[data.FraudIndicator == 1] 
df_non_FraudIndicator = data.loc[data.FraudIndicator == 0]

sns.countplot(x="FraudIndicator",data=data)
plt.title("Count of Fraudulent Payments")
plt.show()
print("Number of normal examples: ",df_non_FraudIndicator.FraudIndicator.count())
print("Number of fradulent examples: ",df_FraudIndicator.FraudIndicator.count())
```

```python
print("Mean feature values per category",data.groupby('Category')[['Amount','FraudIndicator']].mean())
```

```python
import pandas as pd
pd.concat([df_FraudIndicator.groupby('Category')['Amount'].mean(),df_non_FraudIndicator.groupby('Category')['Amount'].mean(),\
           data.groupby('Category')['FraudIndicator'].mean()*100],keys=["Fraudulent","Non-Fraudulent","Percent(%)"],axis=1,\
          sort=False).sort_values(by=['Non-Fraudulent'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,30))
sns.boxplot(x=data.Category,y=data.Amount)
plt.title("Boxplot for the Amount spend in category")
plt.ylim(0,3000)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df_FraudIndicator.Amount, alpha=0.5, label='FraudIndicator',bins=100)
plt.hist(df_non_FraudIndicator.Amount, alpha=0.5, label='nonFraudIndicator',bins=100)
plt.title("Histogram for fraudulent and nonfraudulent payments")
plt.ylim(0,10000)
plt.xlim(0,1000)
plt.legend()
plt.show()
```

```python
print((data.groupby('Age')['FraudIndicator'].mean()*100).reset_index().rename(columns={'Age':'Age','FraudIndicator' : 'Fraud Percent'}).sort_values(by='Age'))
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
print("Unique MerchantID values: ",data.MerchantID.nunique())
data_reduced = data.drop(['MerchantID'],axis=1)
```

```python
data_reduced.columns
```

```python
col_categorical = data_reduced.select_dtypes(include= ['object']).columns
for col in col_categorical:
    data_reduced[col] = data_reduced[col].astype('category')
data_reduced[col_categorical] = data_reduced[col_categorical].apply(lambda x: x.cat.codes)
data_reduced.head
```
